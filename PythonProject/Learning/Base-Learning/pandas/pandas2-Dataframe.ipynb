DataFrame

```python
#创建DataFrame
data = [['Google',10],['Runoob',12],['Wiki',13]]   #定义二维数组
df = pd.DataFrame(data,columns=['Site','Age'])
print(df)
```

```python
#使用字典创建
data = [{'a': 1, 'b': 2},{'a': 5, 'b': 10, 'c': 20},{'a':12,'b':25,'c':10}]
df = pd.DataFrame(data)
print (df)
```

```python
#属性
/*
shape:获取维度
index:获取行索引
columns:获取列索引
head()：查看前几行内容 （默认5行）
tail()：查看后几行内容（默认5行）
*/
```

```python
print(df.shape)
print(df.index)
print(df.columns)
print(df.head(2))
print(df.tail(2))
```

```python
df
```

```python
#索引
#1.直接索引（先列后行，且只能使用标签值）
print(df['a'][2])
print(df['c'][0])
```

```python
#2.loc/iloc函数(loc使用标签索引，iloc使用数字索引，但顺序均为先行后列)
print(df.loc[[0,1],['a','c']])
print(df.iloc[[0,1],[0,2]])
print(df.loc[0,'c'])
print(df.iloc[0,2])
print(df.loc[2])       #返回的是Series格式
print(df.loc[[0,1,2],'a'])  #返回的是Series格式
```

```python
#获取特定行
print(df.loc[0])
print(df.loc[[0,2]])
print(type(df.loc[0]))
```

```python
#获取特定列
print(df['a'])
print(df[['a','c']])
print(df.b)
```

```python
#修改操作
df.loc[0,'c']=100;
print(df)
```

```python
df['c'][0]=10
print(df)
```

```python
df.a[2]=100
print(df)
```

```python
print(df.loc[:,['a','c']])  #第一列和第三列
print(df.loc[:,'a':'c'])   #第一列到第三列
```

```python
#单独选取某些行
#1.
print(df[0:3])
#2.
print(df[0:1])
```

```python
#排序
df2=df.sort_values(by='b',ascending=False) #降序排序
print(df)
print(df2)
#排序后原df不变，返回排序后的副本
```

```python
df3=df.sort_index(axis=0,ascending=False) #按行索引降序排序
print(df)
print(df3)
```

```python
df4=df.sort_index(axis=1,ascending=False) #按行索引降序排序
print(df)
print(df4)
```
