```python
import pandas as pd
pd.__version__
```

```python
#创建DataFrame
##以键值对的形式赋值
mydataset = {
  'sites': ["Google", "Runoob", "Wiki"],
  'number': [1, 2, 3]
}
type(mydataset)  #一开始是字典类型
```

```python
#转化为DataFrame类型
myData=pd.DataFrame(mydataset)  
print(type(myData))
print(myData)
#键变为列标签，对应值列表则变为一列数据
```
