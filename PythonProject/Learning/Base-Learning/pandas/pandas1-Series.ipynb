```python
import pandas as pd
pd.__version__
```

```python
#创建DataFrame
##以键值对的形式赋值
mydataset = {
  'sites': ["Google", "Runoob", "Wiki"],
  'number': [1, 2, 3]
}
type(mydataset)  #一开始是字典类型
```

```python
#转化为DataFrame类型
myData=pd.DataFrame(mydataset)  
print(type(myData))
print(myData)
#键变为列标签，对应值列表则变为一列数据
```

Series数据结构

```python
#Series创建
a=[1,2,3];
myvar=pd.Series(a)
print(myvar)
print(type(myvar))
```

```python
#索引
myvar[1]
```

```python
#指定索引值
a = ["Google", "Runoob", "Wiki"]
myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
```

```python
print(myvar[1])
print(myvar['y'])
#即支持数字索引也支持字符索引
```

```python
#我们也可以使用 key/value 对象，类似字典来创建 Series
sites = {'name': "LKY", 'score': 99.9, 'number': 2023}
myvar = pd.Series(sites)
print(myvar)
print(myvar[0])
print(myvar['score'])
print(myvar["number"])
```
