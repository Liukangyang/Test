#字典

```python
#定义d = {key1 : value1, key2 : value2, key3 : value3 }
#键必须是唯一的，但值则不必。
#值可以取任何数据类型，但键必须是不可变的，如字符串，数字()
```

```python
tinydict1 = { 'name': 'LKY' ,'age':22};
tinydict1
```

```python
len(tinydict1)
```

```python
dict()   #使用内建函数创建空字典
```

```python
#访问字典的值
tinydict1['age']
```

```python
#修改字典
tinydict1['score']=100;  #
tinydict1['age']=23;
tinydict1
```

```python
#删除字典元素
del tinydict1['score']
print(tinydict1)
#tinydict1.clear()  #变为空字典
del tinydict1;
print(tinydict1)  #del 直接删除tinydict1变量
```

```python
#字典特性
#1）不允许同一个键出现两次。创建时如果同一个键被赋值两次，后一个值会被记住
#2）键必须不可变，所以可以用数字，字符串或元组充当，而用列表就不行
tinydict = {['Name']: 'Runoob', 'Age': 7}
```
