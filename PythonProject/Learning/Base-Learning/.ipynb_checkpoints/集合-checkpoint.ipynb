#集合操作

```python
#1.添加元素
set3.add('hello')
set3
```

```python
set3.update(['LKY','LKY2'])
set3
```

```python
set4={1,2,3}
set3.update(set4)
set3
```

```python
#2.移除元素
set3.remove('LKY')
set3
```

```python
set3.discard('LKY2')
set3
```

```python
print(set3.pop())  #随机删除
set3
```

```python
#3.计算集合个数
len(set3)
```

```python
#4.清空集合
set3.clear()
set3   #空集合
```

```python
#5.判断元素是否在集合中存在
print(set3)
's' in set3
```

#集合内置方法

```python
#差集
a={1,2,3,4,5,6,7,8,9};
b={2,4,6,8};
print(a-b)
print(a.difference(b))
a
```

```python
#交集
print(a & b)
print(a.intersection(b))
print(b.intersection(a))
#a.intersection_update(b)
a
```

```python
#并集
print(a | b)
print(a.union(b))
print(b.union(a))
b
```

```python
#判断子集
#1.issubset:判断指定集合是否为该方法参数集合的子集。
b.issubset(a)
```

```python
#2.issuperset() 判断该方法的参数集合是否为指定集合的子集
a.issuperset(b)
```
